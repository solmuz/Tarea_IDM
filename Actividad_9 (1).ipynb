```python
# Incluya aquí su código para importar librerías y datos
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import drive
```

```python
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/responses.csv')
print(df.head())
print(df.info)
print(df.isnull().sum())
```

```python
# Incluya aquí el código para obtener las correlaciones
df_numeric = df.select_dtypes(include=[np.number])
correlation_matrix = df_numeric.corr()

plt.figure(figsize=(20, 15))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', center=0)
plt.title('Matriz de Correlación')
plt.show()
strong_correlations = correlation_matrix[abs(correlation_matrix) > 0.5].stack().reset_index()
strong_correlations = strong_correlations[strong_correlations['level_0'] != strong_correlations['level_1']]  # Eliminar diagonal
strong_correlations = strong_correlations.sort_values(by=0, ascending=False)
print(strong_correlations.head(10))
```

```python
#Código para crear un dataframe solo con los campos de variables Música y Demografía

variables_musica = list(df.columns[0:10])
variables_demografía = list(df.columns[-10:])
lista_variables = variables_musica + variables_demografía
lista_variables
```

```python
data_mus_dem = df[lista_variables]
data_mus_dem
```

```python
# Incluye aquí tu código para hacer el análisis de correlaciones

# Ejemplo: Matriz de correlación de variables de música y demografía
f,ax = plt.subplots(figsize=(10, 8))
numeric_data = data_mus_dem.select_dtypes(include=['float64', 'int64'])
corr = numeric_data.corr()
sns.heatmap(corr,
            mask=np.zeros_like(corr, dtype=bool),  # Corregido np.bool -> bool
            cmap=sns.diverging_palette(220, 10, as_cmap=True),  # Corregido "palette" -> "palette"
            square=True,
            annot=True,
            ax=ax)
plt.show()
```
